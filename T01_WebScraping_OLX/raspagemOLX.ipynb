```
# Importando bibliotecas
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
from google.colab import drive
import pandas as pd
import requests

# Declarando variável cards
cards = []

## Obtendo o HTML e o total de páginas
url = 'https://rn.olx.com.br/rio-grande-do-norte/natal/imoveis/venda/apartamentos?ps=1'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'}

try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    links = str([a for a in soup.find('div', class_="sc-hmzhuo hMZElg sc-jTzLTM iwtnNi")]).split('href=')[-1].split('&amp;')[0].split('=')[-1]
    pages = int(links)
   
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason)
    
## Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    urlPaginas = 'https://rn.olx.com.br/rio-grande-do-norte/natal/imoveis/venda/apartamentos?o=' + str(i + 1) +'&ps=1'
    print(urlPaginas)
    reqPaginas = Request(urlPaginas, headers = headers)
    responsePaginas = urlopen(reqPaginas)
    htmlPaginas = responsePaginas.read().decode('utf-8')
    soup = BeautifulSoup(htmlPaginas, 'html.parser')

    # Obtendo as TAGs de interesse
    anuncios = soup.find('div', {"class": "sc-1fcmfeb-0 FBZzf"}).findAll('li', class_="sc-1fcmfeb-2 juiJqh")

    # Coletando as informações dos CARDS
    for anuncio in anuncios:
        card = {}

        # Valor
        if (anuncio.find('div', {'class': 'aoie8y-0 hRScWw'})):
          valores = str(anuncio.find('div', {'class': 'aoie8y-0 hRScWw'}).findAll('span', class_="sc-ifAKCX eoKYee")).split('="400">')[-1].split('</span>]')[0]
          card['valor'] = valores
          
        #descrição
        if (anuncio.find('div', {'class': 'sc-1iuc9a2-0 eOiLKd'})):
          descricoes = str(anuncio.find('div', {'class': 'sc-1iuc9a2-0 eOiLKd'}).findAll('h2', {'class': 'sc-1iuc9a2-1 daMDOK sc-ifAKCX eKQLlb'})).split('="400">')[-1].split('</h2>]')[0]
          card['descricao'] = descricoes

        #tipo de anuncio e data do anuncio
        if (anuncio.find('p', {'class': 'sc-1iuc9a2-4 hDBjae sc-ifAKCX fWUyFm'})):
          dataPublicacoes = str(anuncio.find('p', {'class': 'sc-1iuc9a2-4 hDBjae sc-ifAKCX fWUyFm'})).split('="400">')[-1].split('<!-- --> <!-- -->- ')[0].split('<!-- --> </p>')[0]
          card['dataPublicacao'] = dataPublicacoes
          tipoPublicacoes = str(anuncio.find('p', {'class': 'sc-1iuc9a2-4 hDBjae sc-ifAKCX fWUyFm'})).split('="400">')[-1].split('<!-- --> <!-- -->- ')[-1].split('</p>')[0]
          card['tipoPublicacao'] = tipoPublicacoes

        # Comodos
        if (anuncio.find('span', {'class': 'sc-1j5op1p-0 lnqdIU sc-ifAKCX eLPYJb'})):
          items = str(anuncio.find('span', {'class': 'sc-1j5op1p-0 lnqdIU sc-ifAKCX eLPYJb'})).split('title=')[-1].split('</span>')[0].split(' | ')
          
          card['items'] = items

        # Adicionando resultado a lista cards
        if (card):
          cards.append(card)
        
print (len(cards))

# Montanto Drive e Criando um DataFrame com os resultados
dataset = pd.DataFrame(cards)
drive.mount('/content/gdrive')
dataset.to_csv('/content/gdrive/My Drive/Colab Notebooks/raspagemOlX.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```
